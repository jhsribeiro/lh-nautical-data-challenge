```python
import pandas as pd
import matplotlib.pyplot as plt

df_vendas = pd.read_csv('../datasets/vendas_2023_2024.csv')
```

```python
print(f'Número de linhas: {df_vendas.shape[0]}')
print(f'Número de colunas: {df_vendas.shape[1]}')
df_vendas.head(10)
```

```python
# Top 10 maiores vendas (possíveis outliers) com IDs
top_10 = df_vendas.nlargest(10, "total")[["id_client", "id_product", "total"]]

print("Top 10 maiores vendas (possíveis outliers):")
# O assign é usado apenas para formatar o print sem alterar o dataframe original
print(top_10.assign(total=top_10["total"].apply(lambda x: f"R$ {x:,.2f}")).to_string(index=False))
```

```python
# Top 10 clientes por valor total de vendas
vendas_cliente = df_vendas.groupby("id_client")["total"].sum().sort_values(ascending=False)
print("Top 10 clientes por valor total de vendas:")
print(vendas_cliente.head(10).apply(lambda x: f"R$ {x:,.2f}").to_string())
```

```python
# Vendas por produto
vendas_produto = df_vendas.groupby("id_product")["total"].sum().sort_values(ascending=False)
print("Top 10 produtos por valor total de vendas:")
print(vendas_produto.head(10).apply(lambda x: f"R$ {x:,.2f}").to_string())
```
