```python
def normalizar_categoria(categoria):
    cat_limpa = str(categoria).strip().lower().replace(' ', '')
    if 'eletr' in cat_limpa:
        return 'Eletrônicos'
    elif 'prop' in cat_limpa:
        return 'Propulsão'
    elif 'anc' in cat_limpa or 'enc' in cat_limpa:
        return 'Ancoragem'
    else:
        return categoria.title()
    
df_produtos['actual_category'] = df_produtos['actual_category'].apply(normalizar_categoria)
df_produtos['actual_category'].unique()
```

```python
# --- PARTE 2: Conversão para Tipo Numérico ---
# Removemos o símbolo monetário e convertemos a coluna price para float
df_produtos['price'] = df_produtos['price'].str.replace('R$ ', '', regex=False).astype(float)
```

```python
# --- PARTE 3: Remoção de Duplicatas ---
# Após a padronização das categorias, linhas que antes eram consideradas "diferentes" 
# apenas pela escrita (ex: códigos 37, 62, 127 e 145) agora são idênticas.
# O drop_duplicates() limpará essas redundâncias.
total_antes = len(df_produtos)
df_produtos = df_produtos.drop_duplicates()
total_depois = len(df_produtos)
```

```python
# --- Resultado para o Gabriel Santos (Tech Lead) ---
print(f"Normalização concluída:")
print(f"- Registros originais: {total_antes}")
print(f"- Registros após remover duplicatas: {total_depois}")
print(f"- Duplicatas removidas: {total_antes - total_depois}")

# Visualização das primeiras linhas normalizadas
print("\nExemplo de dados limpos:")
print(df_produtos.head())
```
