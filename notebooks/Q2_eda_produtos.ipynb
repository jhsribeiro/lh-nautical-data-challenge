> Parte 2 - Converter os valores para o tipo númerico.

```python
# Converter a coluna price para numérica, removendo o símbolo de moeda e convertendo vírgulas para pontos
df_produtos['price'] = df_produtos['price'].str.replace('R$', '', regex=False)
df_produtos['price'] = df_produtos['price'].str.replace(',', '.', regex=False)
df_produtos['price'] = pd.to_numeric(df_produtos['price'])

df_produtos['price'].describe()
```

> Parte 3 - Remover as duplicatas.

```python
# Verificando duplicatas após limpeza
df_produtos.loc[df_produtos.duplicated()]
```

```python
total_antes = len(df_produtos)
print(f"Registros antes: {total_antes}")

# Remover registros duplicados
df_produtos.drop_duplicates(inplace=True)

total_depois = len(df_produtos)
print(f"Registros depois: {total_depois}")
```

```python
# Renomear a coluna 'code' para 'id_product' para melhor clareza e consistência com o nome do arquivo
df_produtos.rename(columns={'code': 'id_product'}, inplace=True)
print(f"Coluna 'code' renomeada para 'id_product'.")
```
