## Questão 3 - Custo de Importação

### Cenário

Além dos dados de produtos, Gabriel percebeu também que o arquivo `custos_importacao.json` tem todos os dados historicos dos preços de compra aninhados num único campo. Para
facilitar a análise dessa informação no banco de dados é necessário melhorarmos isso.

### Premissas obrigatóriao

- Utilize apenas o JSON `custos_importacao.json`
- Utilize obrigatoriamente Python 3

Tarefa:

- Carregue o arquivo JSON e gere um novo arquivo CSV organizando-o de acordo com a definição na imagem abaixo.

| custo_importacao |  |
| --- | --- |
| product_id | integer |
| product_name | text |
| category | text |
| start_date | date |
| usd_date | float |

```python
import pandas as pd
import json
```

```python
with open('../datasets/raw/custos_importacao.json', 'r', encoding='utf-8') as f:
    dados = json.load(f) 
```

```python
# Normalizar os dados
df = pd.json_normalize(
    dados,
    record_path=['historic_data'],
    meta=['product_id', 'product_name', 'category']
)

df.head()
```

```python
# Reorganizar as ordem das colunas
df = df[['product_id', 'product_name', 'category', 'start_date', 'usd_price']]
df.head()
```

```python
print(df.dtypes)
```

```python
# Ajuste dos tipos de dados
df["product_id"] = df["product_id"].astype(int)
df["product_name"] = df["product_name"].astype(str)
df["category"] = df["category"].astype(str)
df["start_date"] = pd.to_datetime(df["start_date"], format="%d/%m/%Y")
df["usd_price"] = df["usd_price"].astype(float)
print(df.dtypes)
```

```python
df['category'].unique()
```

```python
df['category'].value_counts()
```

```python
info_custos = pd.DataFrame({
    "Duplicados": df.duplicated().sum(),
    "Nulos": df.isnull().sum(),
    "Tipo de Dados": df.dtypes,
    "Valores Únicos": df.nunique()
})

info_custos
```

```python
print("          Visão Geral")
print("-"*30)
print(f"Linhas geradas : {len(df)}")
print(f"Produtos únicos: {df['product_id'].nunique()}\n")
print(f"Máximo  : {df['usd_price'].max()}")
print(f"Mínimo  : {df['usd_price'].min()}")
print(f"Média   : {df['usd_price'].mean()}")
print("-"*30)
print()
print("        Tipos dos Dados")
print("-"*30)
print(df.dtypes)
```

```python
# Salvando a base normalizada para uso nas próximas análises
df.to_csv('../datasets/processed/custos_importacao.csv', index=False)
print("\nArquivo salvo → custos_importacao.csv ")
```
