```python
with open('..\datasets\custos_importacao.json', 'r', encoding='utf-8') as f:
    dados = json.load(f)
```

```python
linhas = []
 
# Normalização dos dados para estrutura tabular
for produto in dados: 
    for registro in produto["historic_data"]:
        linhas.append({
            "product_id":   produto["product_id"],
            "product_name": produto["product_name"],
            "category":     produto["category"],
            "start_date":   registro["start_date"]
            "usd_price":    registro["usd_price"],
        })
 
df_custos = pd.DataFrame(linhas)
df_custos.head()
```

```python
# Ajuste dos tipos de dados
df_custos["product_id"] = df_custos["product_id"].astype(int)
df_custos["product_name"] = df_custos["product_name"].astype(str)
df_custos["category"] = df_custos["category"].astype(str)
df_custos["start_date"] = pd.to_datetime(df_custos["start_date"], format="%d/%m/%Y")
df_custos["usd_price"] = df_custos["usd_price"].astype(float)
print(df_custos.dtypes)
```
