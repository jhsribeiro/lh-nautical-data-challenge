```python
import pandas as pd
import json
from datetime import datetime
```

```python
with open('..\datasets\custos_importacao.json', 'r', encoding='utf-8') as f:
    dados = json.load(f) 
```

```python
# Normalizar os dados
df = pd.json_normalize(
    dados,
    record_path=['historic_data'],
    meta=['product_id', 'product_name', 'category']
)

df.head()
```

```python
# Reorganizar as ordem das colunas
df = df[['product_id', 'product_name', 'category', 'start_date', 'usd_price']]
df.head()
```
