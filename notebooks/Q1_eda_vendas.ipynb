* Datas inconsistentes na coluna `sale_date`

```python
# detecta formato sem converter
mask_iso = df_vendas["sale_date"].str.match(r"^\d{4}-\d{2}-\d{2}$")
mask_br  = df_vendas["sale_date"].str.match(r"^\d{2}-\d{2}-\d{4}$")

print(f"Formato YYYY-MM-DD : {mask_iso.sum():,} registros ({mask_iso.mean()*100:.1f}%)")
print(f"Formato DD-MM-YYYY : {mask_br.sum():,}  registros ({mask_br.mean()*100:.1f}%)")
print(f"Outros/inválidos   : {(~mask_iso & ~mask_br).sum()} registros")
```

> Parte 3 - Interpretação
* Possíveis Outliers

```python
# Possíveis outliers em total usando o método IQR
q1 = df_vendas['total'].quantile(0.25)
q3 = df_vendas['total'].quantile(0.75)
iqr = q3 - q1
limite_superior = q3 + 1.5 * iqr

outliers = df_vendas[df_vendas['total'] > limite_superior]

print(f"Q1: R$ {q1:,.2f}")
print(f"Q3: R$ {q3:,.2f}")
print(f"IQR: R$ {iqr:,.2f}\n")
print(f"Limite Superior: R$ {limite_superior:,.2f}\n")
print(f"Outliers identificados pelo método IQR: {len(outliers):,} registros ({len(outliers)/len(df_vendas)*100:.1f}%)")
```

```python
# Top 10 maiores vendas (possíveis outliers) com IDs
top_10 = df_vendas.nlargest(10, "total")[["id_client", "id_product", "total"]]

print("Top 10 maiores vendas (possíveis outliers):")
# O assign é usado apenas para formatar o print sem alterar o dataframe original
print(top_10.assign(total=top_10["total"].apply(lambda x: f"R$ {x:,.2f}")).to_string(index=False))
```

```python
# Top 10 clientes por valor total de vendas
vendas_cliente = df_vendas.groupby("id_client")["total"].sum().sort_values(ascending=False)
print("Top 10 clientes por valor total de vendas:")
print(vendas_cliente.head(10).apply(lambda x: f"R$ {x:,.2f}").to_string())
```

```python
# Vendas por produto
vendas_produto = df_vendas.groupby("id_product")["total"].sum().sort_values(ascending=False)
print("Top 10 produtos por valor total de vendas:")
print(vendas_produto.head(10).apply(lambda x: f"R$ {x:,.2f}").to_string())
```

### Possíveis Outliers na coluna 'total':
O dataset apresenta outliers críticos. As 10 maiores vendas possuem valores na casa dos milhões (ex: R$ 2.222.973,00), o que é desproporcional à média. Mais grave ainda: os valores de topo repetem-se exatamente até os centavos, sugerindo que não são vendas orgânicas, mas sim erros de processamento ou registros duplicados que inflam artificialmente o faturamento.

### Qualidade dos Dados (Nulos e Inconsistências):
A qualidade é baixa devido a inconsistências estruturais:

* **Datas**: Há uma mistura perigosa de formatos (`AAAA-MM-DD` e `DD-MM-AAAA`), o que impede qualquer análise temporal confiável sem conversão prévia.
* **Valores Nulos**: Não há valores nulos.
* **Unicidade**: A repetição de IDs e valores exatos em vendas de alto valor indica falta de integridade nos registros.

### O Dataset está pronto para análise?
**Não**. O dataset exige tratamento prévio rigoroso antes de qualquer tomada de decisão. Utilizá-lo como está resultaria em uma visão distorcida da empresa, superestimando o faturamento e errando as projeções de datas.

> **Recomendação**: É necessário realizar uma limpeza (Data Cleaning) para padronizar datas, remover ou investigar duplicatas de alto valor e tratar os valores nulos antes de prosseguir.

```python
import sqlite3

# Criar uma ligação com uma base de dados SQLite na memória do Python
conn = sqlite3.connect(':memory:')

# Enviar o DataFrame para a base de dados como uma tabela SQL
df_vendas.to_sql('vendas_2023_2024', conn, index=False)

query_q1 = """
SELECT 
    COUNT(*) AS total_linhas,
    MIN(sale_date) AS data_minima,
    MAX(sale_date) AS data_maxima,
    MIN(total) AS valor_minimo,
    MAX(total) AS valor_maximo,
    ROUND(AVG(total), 2) AS valor_medio
FROM vendas_2023_2024;
"""
print("--- Resultado da Questão 1.1 (Métricas) ---")
display(pd.read_sql_query(query_q1, conn))

query_colunas = "SELECT count(*) as total_colunas FROM pragma_table_info('vendas_2023_2024');"
print("\n--- Resultado da Questão 1.1 (Colunas) ---")
display(pd.read_sql_query(query_colunas, conn))
```
